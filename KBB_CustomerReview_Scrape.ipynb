```python
import requests
import csv
import time
import pandas as pd
from bs4 import BeautifulSoup
```

```python
segmentdf = pd.read_excel('VehicleSegments.xlsx').drop('Unnamed: 0', axis = 1)
year_list = ['2020', '2021', '2022', '2023']
```

```python
#segmentdf = segmentdf[0:5]
```

```python
data = []
t = time.time()
data = []
reviews_df = pd.DataFrame(columns = ["Segment", "Make", "Model", "Year", "Reviews"])
for j in year_list:
    #iterate through models
    for k in segmentdf['Model']:
        print(time.time() -t)
        print(j,k)
        try:
            counter = 0
            make_url = segmentdf.loc[segmentdf['Model'] == k]['Make'].tolist()[0]
            segment = segmentdf.loc[segmentdf['Model'] == k]['Segment'].tolist()[0]
            if len(k.split(' ')) >= 2:
                kr = k.replace(' ', '-')
            else:
                kr = k
            url = 'https://www.kbb.com/'+make_url+'/'+kr+'/'+j+'/consumer-reviews'
            page = requests.get(url)
            if page.status_code == 200:
                #while url:
                soup = BeautifulSoup(page.content, 'html.parser')
                title = soup.find_all('div', class_='css-1c7qqqr')
                reviews = soup.find_all('p', class_='css-25a2lr emgezi80')
    
                for i in range(len(title)):
                    data.append((segment, make_url, k, j, title[i].get_text() + ' '+ reviews[i].get_text()))
                    time.sleep(1)

            else:
                continue
        except Exception as e:
            print(e)
            continue
```

```python
df = pd.DataFrame(data, columns = ["Segment", "Make", "Model", "Year", "Review"])
```

```python
df_final = df.drop_duplicates()
df_final = df_final.dropna()
df_final.to_csv('KBBCustomerReviews.csv')
```
