```python
import requests
import csv
import time
import pandas as pd
from bs4 import BeautifulSoup
```

```python
segmentdf = pd.read_excel('EV_VehicleSegments.xlsx').drop('Unnamed: 0', axis = 1)
year_list = ['2020', '2021', '2022', '2023']
```

```python
#underscore replacement is required based on cars urls
segmentdf['Model'] = segmentdf['Model'].str.replace('-', '_')
```

```python
data = []
t = time.time()
data = []
reviews_df = pd.DataFrame(columns = ["Segment", "Make", "Model", "Year", "Reviews"])
for j in year_list:
    #iterate through models
    for k in segmentdf['Model']:
        print(time.time() -t)
        print(j,k)
        try:
            counter = 0
            make_url = segmentdf.loc[segmentdf['Model'] == k]['Make'].tolist()[0]
            segment = segmentdf.loc[segmentdf['Model'] == k]['Segment'].tolist()[0]
            if len(k.split(' ')) >= 2:
                #underscore is required based on cars urls
                kr = k.replace(' ', '_')
            else:
                kr = k
            url = 'https://www.cars.com/research/'+make_url+'-'+kr+'-'+j+'/consumer-reviews?page_size=200'
            print(url)
            page = requests.get(url)
            if page.status_code == 200:
                #while url:
                soup = BeautifulSoup(page.content, 'html.parser')
                    # title
                title = soup.find_all('h3', class_='sds-heading--7 title')
                    #review sub-title
                    #sub_title = soup.find_all('div', class_='review-byline review-section')
                    #review content
                reviews = soup.find_all('p', class_='review-body')
                    #create a list of tuples with the data
    
                for i in range(len(title)):
                    data.append((segment, make_url, k, j, title[i].get_text() + ' '+ reviews[i].get_text()))
                    time.sleep(1)
                    
                    #next_link = soup.find("a",{"rel":"next"})
                    #if next_link:
                    #    url = next_link["href"]
                    #else:
                    #url = None
            else:
                continue
        except Exception as e:
            print(e)
            continue
```

```python
df = pd.DataFrame(data, columns = ["Segment", "Make", "Model", "Year", "Review"])
```
