```python
import requests
import csv
import time
import itertools
from bs4 import BeautifulSoup
import pandas as pd
pd.set_option('display.max_rows', 200)
```

```python
segmentdf = pd.read_excel('VehicleSegments.xlsx').drop('Unnamed: 0', axis = 1)
```

```python
year_list = ['2020', '2021', '2022', '2023']
#for i in segmentdf['Model']:
#    print(segmentdf.loc[segmentdf['Model'] == i]['Make'].tolist()[0])
#    print(segmentdf.loc[segmentdf['Model'] == i]['Segment'].tolist()[0])
```

```python
t = time.time()
data = []
reviews_df = pd.DataFrame(columns = ["Segment", "Make", "Model", "Year", "Reviews"])
#iterate through years
comp = 1
for j in year_list:
    #iterate through models
    for k in segmentdf['Model']:
        print(comp)
        print(time.time() -t)
        print(j,k)
        comp = comp + 1
        try:
            counter = 0
            #page navigation for single model 
            for i in itertools.count():
                counter = counter+1
                make_url = segmentdf.loc[segmentdf['Model'] == k]['Make'].tolist()[0]
                url = 'https://www.edmunds.com/'+make_url+'/'+k+'/'+str(j)+'/consumer-reviews/?pagenum='+str(counter)+''
                #print(url)
                headers = {"User-Agent": "Mozilla/5.0 (X11; CrOS x86_64 12871.102.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.141 Safari/537.36"}
                page = requests.get(url, timeout = 5, headers = headers)
                #print('Yes')
                if page.status_code == 200:
                    #print('Ok')
                    segment = segmentdf.loc[segmentdf['Model'] == k]['Segment'].tolist()[0]
                    soup = BeautifulSoup(page.content, 'html.parser')
                    reviews=soup.find_all('p')
                    #print('Parsed for one Page')
                    #For 2022 onwards use below
                    #if "Receive pricing updates, shopping tips &amp; more!" in str(reviews[0]):
                    #Before 2022 use below
                    if int(j) < 2022:
                        if "Download the Edmunds app" in str(reviews[4]):
                            break
                        for i in range(len(reviews)):
                            data.append((segment, make_url, k, j, reviews[i].get_text()))
                    #print('Appended for 1 page')
                    #print(counter)
                    #print(data[-3:])
                    else:
                        if "Receive pricing updates, shopping tips &amp; more!" in str(reviews[0]):
                            break
                        for i in range(len(reviews)):
                            data.append((segment, make_url, k, j, reviews[i].get_text()))
                    time.sleep(2)
                else:
                    break
        except:
            continue
```

```python
df = pd.DataFrame(data, columns = ["Segment", "Make", "Model", "Year", "Review"])
```

```python
df.head(10)
```

```python
to_del = df.loc[(df["Review"] == "Receive pricing updates, shopping tips & more!") | (df["Review"] == "Download the Edmunds app") | (df["Review"] == "Has Your Car's Value Changed?") | (df["Review"] == "See Edmunds pricing data") | (df["Review"] == "Used car values are constantly changing. Edmunds lets you track your vehicle's value over time so you can decide when to sell or trade in.")].index
df2 = df.drop(to_del)
df2.drop_duplicates(subset = "Review", keep = "first", inplace = True)
df2 = df2.reset_index()
df2.drop('index', axis = 1, inplace = True)
```

```python
#This was scraped earlier through Legacy code which only considered Nissan Vehicles. (Can be found in Legacy Code Dump)
#Maxima and Versa were missing in the Segment dataset provided by Nissan hence, these are readded. 
edm1 = pd.read_csv('Edmunds_CustomerReviews_Clean.csv', encoding = 'utf-16')
edm2 = pd.read_csv('Edmunds_CustomerReviews_Old.csv', encoding = 'utf-16')
```

```python
edm = pd.concat([edm1, edm2])
edm.drop_duplicates(subset = "Review", keep = "first", inplace = True)
edm.drop(["Unnamed: 0", "Type"], axis = 1, inplace = True)
```

```python
df2.loc[df2["Make"] == "Nissan"]["Model"].value_counts()
```

```python
edm["Model"].value_counts()
```

```python
missing_v = edm.loc[(edm["Model"] == "Nissan Maxima") | (edm["Model"] == "Nissan Versa")]
missing_v['Make'] = missing_v['Model'].str.split().str.get(0)
missing_v['Model'] = missing_v['Model'].str.split().str.slice(1).str.join(' ')
def new_col(row):
    if row['Model'] == "Versa":
        return 'Compact'
    else:
        return 'Lower Midsize'
missing_v['Segment'] = missing_v.apply(lambda row: new_col(row), axis=1)
missing_v = missing_v[['Segment', 'Make', 'Model', 'Year', 'Review']]
```

```python
df3 = pd.concat([df2, missing_v])
df3 = df3.reset_index()
df3.drop('index', axis = 1, inplace = True)
```

```python
df3.to_csv('Edmunds_CustomerReviews.csv', encoding = "utf-16")
```
